```python
import pandas as pd

data = pd.read_csv('../input/digit-recognizer/train.csv')
y = pd.get_dummies(data['label']).to_numpy(dtype='float32')
data.pop(item='label')
X = data.to_numpy(dtype='float32')
X = X/255.0
X = X.reshape((-1,28,28,1))
X.shape, y
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPool2D

model = tf.keras.models.Sequential()

model.add(Conv2D(32, (3,3), input_shape=(28,28,1), name='conv1'))
model.add(MaxPool2D(pool_size=(2,2), name='pool1'))

model.add(Conv2D(32, (3,3), name='conv2'))
model.add(MaxPool2D(pool_size=(2,2), name='pool2'))

model.add(Flatten())
model.add(Dense(64, activation=tf.nn.relu))
model.add(Dense(10, activation=tf.nn.softmax))

model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=['accuracy'])
```

```python
history = model.fit(X, y, epochs=3, validation_split=0.15, batch_size=20)
```

```python
pd.DataFrame(history.history).plot()
```

```python
data = pd.read_csv('../input/digit-recognizer/test.csv')
X_test = data.to_numpy(dtype='float32')
X_test = X_test/255.0
X_test = X_test.reshape((-1,28,28,1))
X_test.shape
```

```python
y_pred = model.predict(X_test)
y_pred, y_pred.shape
```

```python
pred_labels = pd.DataFrame(y_pred).idxmax(axis=1)
d =  { 'ImageId' : [i for i in range(1,28001)], 'Label' : pred_labels.values}
result = pd.DataFrame(d)
result.to_csv('submission.csv', index=False)
```
